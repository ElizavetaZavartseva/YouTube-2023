# Creating new dataframes

## New table "country"

```python
country_df = youtube_df[['country', 'abbreviation', 
                         'gross_tertiary_education_enrollment_(%)', 'population', 
                         'unemployment_rate', 'urban_population', 'latitude', 'longitude']].drop_duplicates().reset_index(drop=True)
country_df
```

```python
country_df.dropna(axis=0, inplace = True, how='all')

#Deleting rows with null-values
```

```python
country_df.reset_index(drop=True, inplace=True)

#Reseting index
```

```python
# country_df.index.names = ['country_id']
```

```python
country_df
```

```python
country_df.index = np.arange(1, len(country_df)+1)
```

```python
country_df.index.names = ['country_id']
```

```python
country_df
```

```python
country_df.to_excel("countries.xlsx")
```

```python
country_df.to_csv("countries.csv")
```

## New table "category" 

```python
category_df = youtube_df.groupby('category')[["subscribers", "video_views"]].sum()
```

```python
category_df
```

```python
category_df.reset_index(inplace=True)
```

```python
category_df
```

```python
category_df.index = np.arange(1, len(category_df)+1)
```

```python
category_df
```

```python
category_df.index.names = ['category_id']
```

```python
category_df
```

```python
category_df['popularity_rating'] = category_df['video_views']/category_df['subscribers']
```

```python
category_df['popularity_rating'] = category_df['popularity_rating'].round(2)
```

```python
category_df
```

```python
category_df.to_excel("categories.xlsx")
```

```python
category_df.to_csv("categories.csv")
```
