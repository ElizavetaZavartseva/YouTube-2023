# Creating new dataframes

## New table "country"

```python
country_df = youtube_df[['country', 'abbreviation', 
                         'gross_tertiary_education_enrollment_(%)', 'population', 
                         'unemployment_rate', 'urban_population', 'latitude', 'longitude']].drop_duplicates().reset_index(drop=True)
country_df
```

```python
country_df.dropna(axis=0, inplace = True, how='all')

#Deleting rows with null-values in all columns
```

```python
country_df.reset_index(drop=True, inplace=True)
country_df.index = np.arange(1, len(country_df)+1)
country_df.index.names = ['country_id']

country_df = country_df.reset_index()

#Adding new column "country_id" and reseting index
```

```python
country_df
```

## New table "category" 

```python
category_df = youtube_df.groupby('category')[["subscribers", "video_views"]].sum()
```

```python
category_df.reset_index(inplace=True)
category_df.index = np.arange(1, len(category_df)+1)
category_df.index.names = ['category_id']
category_df = category_df.reset_index()

#Adding new column "category_id" and reseting index
```

```python
category_df
```

```python
category_df['popularity_rating'] = category_df['video_views']/category_df['subscribers']
category_df['popularity_rating'] = category_df['popularity_rating'].round(2)

#Adding new column "popularity_rating"
```

```python
category_df
```

## New table "channels"

```python
channels_df = youtube_df[['rank', 'youtuber', 
                         'subscribers', 'video_views', 
                         'category', 'uploads', 'country', 'video_views_for_the_last_30_days',
                         'lowest_monthly_earnings', 'highest_monthly_earnings', 
                         'lowest_yearly_earnings', 'highest_yearly_earnings',
                         'subscribers_for_last_30_days', 'created_year',
                         'created_month', 'created_date']].drop_duplicates().reset_index(drop=True)
```

```python
channels_df.reset_index(drop=True, inplace=True)
channels_df.index = np.arange(1, len(channels_df)+1)
channels_df.index.names = ['channel_id']
channels_df = channels_df.reset_index()

#Adding new column "channel_id" and reseting index
```

```python
channels_df = pd.merge(channels_df, category_df, on='category', how='left')

#Merging 2 DFs in ONE df by column "category" (to add foreign key)
```

```python
channels_df.drop('subscribers_y', axis=1, inplace=True)
channels_df.drop('video_views_y', axis=1, inplace=True)
channels_df.drop('popularity_rating', axis=1, inplace=True)

#Removing unneeded colimns
```

```python
channels_df = pd.merge(channels_df, country_df, on='country', how='left')

#Merging 2 DFs in ONE df by column "country" (to add foreign key)
```

```python
channels_df.drop('abbreviation', axis=1, inplace=True)
channels_df.drop('gross_tertiary_education_enrollment_(%)', axis=1, inplace=True)
channels_df.drop('population', axis=1, inplace=True)
channels_df.drop('unemployment_rate', axis=1, inplace=True)
channels_df.drop('urban_population', axis=1, inplace=True)
channels_df.drop('latitude', axis=1, inplace=True)
channels_df.drop('longitude', axis=1, inplace=True)
channels_df.drop('category', axis=1, inplace=True)
channels_df.drop('country', axis=1, inplace=True)

#Removing unneeded colimns
```

```python
channels_df.loc[channels_df["created_year"].isna()]

#Checking Null-values in column "created_year"
```

```python
channels_df.columns = channels_df.columns.str.replace('subscribers_x', 'subscribers') 
channels_df.columns = channels_df.columns.str.replace('video_views_x', 'video_views') 

#Changing names of columns
```

## Table: categories

```python
import mysql.connector
```

```python
def SQL_INSERT_STATEMENT_FROM_DATAFRAME(SOURCE, TARGET):
    sql_texts = []
    for index, row in SOURCE.iterrows():       
        sql_texts.append('INSERT INTO '+TARGET+' ('+ str(', '.join(SOURCE.columns))+ ') VALUES '+ str(tuple(row.values)))        
    return sql_texts

def SQL_CREATE_STATEMENT_FROM_DATAFRAME(SOURCE, TARGET):
    # SQL_CREATE_STATEMENT_FROM_DATAFRAME(SOURCE, TARGET)
    # SOURCE: source dataframe
    # TARGET: target table to be created in database

    sql_text = pd.io.sql.get_schema(SOURCE.reset_index(), TARGET)   
    return sql_text

user = 'root'
password = 'Qwerty123'
host = 'localhost'
database = 'youtube_db'

db = mysql.connector.connect(
    user=user,
    password=password,
    host=host,
    database=database
)

# from sqlalchemy import create_engine
# engine = create_engine(f"mysql://root:Qwerty123@localhost/youtube_db")

insert_statements_list = SQL_INSERT_STATEMENT_FROM_DATAFRAME(category_df, "categories")
q = SQL_CREATE_STATEMENT_FROM_DATAFRAME(category_df, "categories")
# print(q)

sql = '''
        SET FOREIGN_KEY_CHECKS = 0;
        DROP TABLE IF EXISTS `categories`;
        CREATE TABLE `categories` (
          `category_id` int NOT NULL,
          `category` varchar(45) DEFAULT NULL,
          `subscribers` bigint(20) DEFAULT NULL,
          `video_views` bigint(20) DEFAULT NULL,
          `popularity_rating` decimal(20,2) DEFAULT NULL,
          PRIMARY KEY (`category_id`)
        ) ENGINE=InnoDB DEFAULT CHARSET=utf8mb3;
        ''' + "; ".join(insert_statements_list)

with db.cursor(buffered=True, dictionary=True) as c: 
    executables = c.execute(sql, multi=True)
    _ = [item for item in executables]
    db.commit() 
print("cool!")
```

## Table: countries

```python
# country_df_1["gross_tertiary_education_enrollment"] = country_df_1["gross_tertiary_education_enrollment_(%)"]
# country_df_1.drop(['B', 'C'], axis=1)
insert_statements_list = SQL_INSERT_STATEMENT_FROM_DATAFRAME(country_df, "countries")
# print(insert_statements_list)

sql = '''
        DROP TABLE IF EXISTS `countries`;
        CREATE TABLE `countries` (
          `country_id` int NOT NULL,
          `country` varchar(45),
          `abbreviation` varchar(45),
          `gross_tertiary_education_enrollment` decimal(20,2) DEFAULT NULL,
          `population` decimal(20) DEFAULT NULL,
          `unemployment_rate` decimal(20,2) DEFAULT NULL,
          `urban_population` decimal(20) DEFAULT NULL,
          `latitude` decimal(20,10) DEFAULT NULL,
          `longitude` decimal(20,10) DEFAULT NULL,
          PRIMARY KEY (`country_id`)
        ) ENGINE=InnoDB DEFAULT CHARSET=utf8mb3;
        ''' + "; ".join(insert_statements_list)

sql = sql.replace("gross_tertiary_education_enrollment_(%)", "gross_tertiary_education_enrollment").replace("nan", "NULL")

with db.cursor(buffered=True, dictionary=True) as c: 
    executables = c.execute(sql, multi=True)
    _ = [item for item in executables]
    db.commit() 
print("cool!")
```

## Table: channels

```python
a = SQL_CREATE_STATEMENT_FROM_DATAFRAME(channels_df, "channels")
print(a)
```

```python
# country_df_1["gross_tertiary_education_enrollment"] = country_df_1["gross_tertiary_education_enrollment_(%)"]
# country_df_1.drop(['B', 'C'], axis=1)
insert_statements_list = SQL_INSERT_STATEMENT_FROM_DATAFRAME(channels_df, "channels")
# print(insert_statements_list)

sql = '''
        DROP TABLE IF EXISTS `channels`;
        CREATE TABLE `channels` (
              `channel_id` int NOT NULL,
              `rank` int NOT NULL,
              `youtuber` varchar(255),
              `subscribers` bigint(20) DEFAULT NULL,
              `video_views` decimal(20,0) DEFAULT NULL,
              `uploads` bigint(20) DEFAULT NULL,
            
              `video_views_for_the_last_30_days`  decimal(20,0) DEFAULT NULL,
              `lowest_monthly_earnings`  decimal(20,0) DEFAULT NULL,
              `highest_monthly_earnings`  decimal(20,0) DEFAULT NULL,
              `lowest_yearly_earnings` decimal(20,0) DEFAULT NULL,
              `highest_yearly_earnings` decimal(20,0) DEFAULT NULL,
              `subscribers_for_last_30_days` decimal(20,0) DEFAULT NULL,
              `created_year` decimal(20,0),
              `created_month` varchar(45),
              `created_date` decimal(20,0),

              `category_id` int DEFAULT NULL,
              `country_id` int DEFAULT NULL,

              
              PRIMARY KEY (`channel_id`),
              INDEX `fk_channels_countries_idx` (`country_id` ASC) VISIBLE,
              INDEX `fk_channels_categories1_idx` (`category_id` ASC) VISIBLE,
              CONSTRAINT `fk_channels_countries`
                FOREIGN KEY (`country_id`)
                REFERENCES `youtube_db`.`countries` (`country_id`)
                ON DELETE NO ACTION
                ON UPDATE NO ACTION,
              CONSTRAINT `fk_channels_categories1`
                FOREIGN KEY (`category_id`)
                REFERENCES `youtube_db`.`categories` (`category_id`)
                ON DELETE NO ACTION
                ON UPDATE NO ACTION

        ) ENGINE=InnoDB DEFAULT CHARSET=utf8mb3;
        ''' + "; ".join(insert_statements_list)



# print(sql)

sql = sql.replace("rank", "rank_of_channel").replace("nan", "NULL")

with db.cursor(buffered=True, dictionary=True) as c: 
    executables = c.execute(sql, multi=True)
    _ = [item for item in executables]
    db.commit() 
print("cool!")
```

# Vizualization

## Hypothesis: The creation of YouTube channels is more prevalent in countries with higher unemployment rates.

```python
db = mysql.connector.connect(
    user=user,
    password=password,
    host=host,
    database=database
)

df_numbers_per_country = pd.read_sql('SELECT * FROM numbers_per_country', con=db)

df_numbers_per_country = pd.DataFrame(df_numbers_per_country)
```

```python
df_numbers_per_country
```

```python
#Melt the DataFrame to combine 'median_income' and 'crime_rate' into one variable
melted_df = pd.melt(df_numbers_per_country, id_vars=['country'], value_vars=['numbers_of_channels', 'unemployment_rate'],
                    var_name='variable', value_name='value')

melted_df = melted_df.sort_values(by='value', ascending=False)

# Plotting with Seaborn
plt.figure(figsize=(12, 8))
colors = sns.color_palette(['#ff0000', '#000000'])
sns.barplot(y='value', x='country', hue='variable', data=melted_df, palette=colors)

# Adding labels and title
plt.xlabel('Country')
plt.ylabel('Values')
plt.title('Numbers of channels and Unemployment rate')

# Display the plot
plt.legend(title='Variable')
plt.show()
```

## Hypothesis: Channels with higher uploads have faster subscriber growth.

```python
db = mysql.connector.connect(
    user=user,
    password=password,
    host=host,
    database=database
)

uploads_subscribers_view_df = pd.read_sql('SELECT * FROM uploads_subscribers_view', con=db)

uploads_subscribers_view_df = pd.DataFrame(uploads_subscribers_view_df)
```

```python
uploads_subscribers_view_df
```

```python
# Plotting the relationship between uploads and subscribers
plt.figure(figsize=(10, 6))
plt.scatter(uploads_subscribers_view_df['avg_uploads'], uploads_subscribers_view_df['avg_subscribers'], s=100, c='b', alpha=0.5)
plt.title('Relationship between Uploads and Subscribers by Category')
plt.xlabel('Uploads (average)')
plt.ylabel('Subscribers (average)')
plt.grid(True)
plt.show()
```

## Hypothesis: Videos with higher video views also have higher subscriber counts.

```python
db = mysql.connector.connect(
    user=user,
    password=password,
    host=host,
    database=database
)

views_subscribers_view_df = pd.read_sql('SELECT * FROM views_subscribers_new', con=db)

views_subscribers_view_df = pd.DataFrame(views_subscribers_view_df)
```

```python
views_subscribers_view_df
```

```python
# Plotting the relationship between video views and subscribers
plt.figure(figsize=(10, 6))
sns.scatterplot(x='video_views', y='subscribers', data=views_subscribers_view_df, s=100, color='skyblue', edgecolor='black', alpha=0.75)
plt.title('Relationship between Video Views and Subscribers')
plt.xlabel('Video Views')
plt.ylabel('Subscribers')
plt.grid(True)
plt.show()

# Calculating correlation
correlation = views_subscribers_view_df['video_views'].corr(views_subscribers_view_df['subscribers'])
print(f"Correlation between Video Views and Subscribers: {correlation:.2f}")
```
